## Clean up the data
Did you notice anything unusual about the "house" amounts?

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Summarize the Data
 What is average house price at the two levels of leave?

 Conclusion?  Stayers have substantially more expensive homes.

```python
df_clean[['house','leave']].groupby('leave').mean().round(2)
```

## Additional Data Preparation

 Assignment Tasks
- Add at least two additional steps to clean or filter the dataset

```python
df_clean.fillna(df_clean.mean(), inplace=True)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_cols = ['income', 'overage', 'leftover', 'house', 'handset_price']
df_clean[numerical_cols] = scaler.fit_transform(df_clean[numerical_cols])
```

## Data Visualization

 Assignment Tasks
- Add at least two charts to show a summary of any major trends

```python
import matplotlib.pyplot as plt

# Histogram for Income
plt.figure(figsize=(10, 6))
plt.hist(df_clean['income'], bins=30, color='skyblue', edgecolor='black')
plt.title('Income Distribution among Customers')
plt.xlabel('Income')
plt.ylabel('Number of Customers')
plt.grid(True)
plt.show()

# Histogram for Handset Price
plt.figure(figsize=(10, 6))
plt.hist(df_clean['handset_price'], bins=30, color='salmon', edgecolor='black')
plt.title('Handset Price Distribution')
plt.xlabel('Handset Price')
plt.ylabel('Number of Customers')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.boxplot([df_clean[df_clean['leave'] == 'LEAVE']['house'], df_clean[df_clean['leave'] == 'STAY']['house']], labels=['Leave', 'Stay'])
plt.title('Comparison of House Prices Between Customers Who Leave and Stay')
plt.ylabel('House Price')
plt.grid(True)
plt.show()
```
